```python
import pandas as pd
import matplotlib.pyplot as plt

# Função para criar um DataFrame com os meses como índices
def criar_dataframe_meses():
    meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
    return pd.DataFrame(index=meses, columns=[str(ano) for ano in range(2020, 2025)])

# Função para preencher o DataFrame com as vendas de cada mês
def preencher_dataframe_vendas(df, dados, ano, produto):
    for mes in df.index:
        vendas_mes = dados[(dados['Nome'] == produto)][mes].sum()
        df.loc[mes, str(ano)] = vendas_mes

# Ordenar os dados de vendas de 2023 por total e pegar os 100 produtos mais vendidos
dados_ordenados = vendas2023.sort_values(by='Total', ascending=False)
produtos_mais_vendidos = dados_ordenados['Nome'].head(100).tolist()

# Criar uma lista de meses que se repete para cada ano
meses_continuos = [f'{mes} {ano}' for ano in range(2020, 2025) for mes in criar_dataframe_meses().index]

# Grupos de produtos
grupos_produtos = [produtos_mais_vendidos[i:i+25] for i in range(0, 100, 25)]

# Criar 4 subplots para os gráficos de demanda
fig, axs = plt.subplots(4, 1, figsize=(16, 60), sharex=True, sharey=True)
axs = axs.ravel()

for i, grupo in enumerate(grupos_produtos):
    axs[i].set_title(f'Gráfico {i+1}')
    for produto in grupo:
        df_meses = criar_dataframe_meses()  # Cria um DataFrame novo para cada produto
        for ano, dados in zip(range(2020, 2025), [vendas2020, vendas2021, vendas2022, vendas2023, vendas2024]):
            preencher_dataframe_vendas(df_meses, dados, ano, produto)  # Passa o nome do produto para a função preencher_dataframe_vendas
        vendas_continuas = [vendas for ano in df_meses.columns for vendas in df_meses[ano]]  # Atualiza a lista de vendas continuas
        axs[i].plot(meses_continuos, vendas_continuas, marker='o', label=produto)
    axs[i].legend()
    axs[i].set_xlabel('Mês')
    axs[i].set_ylabel('Quantidade de Vendas')
    axs[i].tick_params(axis='x', rotation=45)
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
